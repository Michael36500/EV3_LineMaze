```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
read = open("najeto_na_kolech2.txt")
read = read.readlines()
```

```python
cisla = np.array([], dtype="int32")
```

```python
for x in read:
    y = x.split()
    a = int(y[0])
    b = int(y[1])
    # cisla.append([a, b])
    cisla = np.append(cisla, (a,b))
a, b, x, y = None, None, None, None
```

```python
cisla = cisla.reshape((int(cisla.shape[0]/2) ,2))
```

```python
mcis = np.array([], dtype="int32")

for x in range(len(cisla)):
    # print(x)
    if x % 4 == 0:
        # print(x)
        mcis = np.append(mcis, cisla[x])
mcis = mcis.reshape((int(mcis.shape[0]/2) ,2))
```

```python
idk = cisla[:, 1]
idk2 = idk.copy()
for x in range(len(idk)-1):
    x += 1
    idk[x] = idk2[x] - idk2[x - 1]

for x in range(len(idk)):
    idk[x] = round(idk[x] / 90, 0)
```

```python
for x in range(len(idk)):
    if idk[x] < 0:
        idk[x] = 0
```
